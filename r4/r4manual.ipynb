```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
def displaydata(result, mult, inhab, player=-1):
    result[mult] = [inhab]
    result[mult].append(float(mult)/inhab)
    if player != -1:
        result[mult].append(player)
        result[mult].append(mult/(inhab+player))
    #print(result)
def graphthis(r2):
    sorted_data = sorted(r2.items())
    mults = [k for k, _ in sorted_data]
    inhabs = [v[0] for _, v in sorted_data]
    mult_per_inhab = [v[1] for _, v in sorted_data]
    players = [v[2] for _, v in sorted_data]
    mult_per_total = [v[3] for _, v in sorted_data]

    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(mults, inhabs, 'g-o', label='inhab')
    plt.plot(mults, mult_per_inhab, 'b-o', label='percieved')
    plt.plot(mults, players, 'r--s', label='player')
    plt.plot(mults, mult_per_total, 'm--s', label='actual')

    plt.title('Metrics vs Mult')
    plt.xlabel('Mult')
    plt.ylabel('Value')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
r2 = dict()
displaydata(r2, 10, 1, 1)
displaydata(r2, 80, 6, 18.2)
displaydata(r2, 37, 3, 5.1)
displaydata(r2, 17, 1, 7.5)
displaydata(r2, 90, 10, 11.8)

displaydata(r2, 31, 2, 7)
displaydata(r2, 50, 4, 8.5)
displaydata(r2, 20, 2, 1.6)
displaydata(r2, 73, 4, 24)
displaydata(r2, 89, 8, 15.2)
#print("Mult, inhab, perceived value, actual percent, value")
totalinhab =0 
totalval = 0

for mult,datalist in r2.items():
    totalval += mult
    totalinhab += datalist[0]
graphthis(r2)
print(round(totalval/(100 + totalinhab),2))
```
